```python
#Section 4.1#
secret = 4
guess = 9

if guess < secret:
    print("too low")

if guess > secret:
    print("too high")

if guess == secret:
    print("just right")
```

```python
#Section 4.2#
small = True
green = True

if small and not green:
    print("Cherry")
elif small and green:
    print("Pea")
elif not small and green:
    print("Watermelon")
else:
    print("Pumpkin")

```

```python
#Section 6.2#
guess_me = 7
number = 1

while True: 
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:
        print("oops")
        break 
    number += 1
```

```python
#Section 6.3#
guess_me = 5

for number in range(10):
    if number > guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:
        print("oops")
        break 
```
